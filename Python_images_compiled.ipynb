```python
from plotnine import *
from plotnine.data import mtcars
import pandas as pd
from mizani.formatters import percent_format
```

```python
age= pd.read_csv("age_deaths_covid.csv")

(ggplot(age, aes(x= "Age Group", y= "7-Day Average", fill= "Age Group"))
+scale_y_continuous(breaks=range(0, 10, 1))
+geom_boxplot()
+theme(figure_size=(16, 8))
+labs(
x= "Age Range", 
y= "7-day Average Deaths",
title= "Age Range and Covid Deaths")
) 
```

```python
age['Date of Death']= pd.to_datetime(age['Date of Death'])
(ggplot(age, aes(x= "Date of Death", y="7-Day Average", group= "Age Group"))
+geom_line(aes(color="Age Group"))
+scale_x_date()
+labs(x= "Date", y= "7-Day Death Average")
+theme(figure_size=(16, 8))
+labs(title= "7-day Death Average vs. Date")
)
```

```python
test= pd.read_csv("Testing_Sequencing1.csv")
(ggplot(test, mapping=aes(x="Lineage", y="Number", fill= "Lineage"))
+geom_boxplot()
+scale_y_continuous(breaks=range(0, 6500, 500))
+theme(figure_size=(16, 8))
+labs(title= "Covid Variants and Weekly Number of Infections")
)
```

```python
test["Percentage"] = pd.to_numeric(test["Percentage"], errors="coerce")
(ggplot(data= test, mapping=aes(x="Lineage", y="Percentage", fill="Lineage"))
+geom_boxplot()
+scale_y_continuous(breaks=range(0, 105, 5))
+labs(title="Percentage of Cases Caused by Each Variant")
+theme(figure_size=(16, 8))
)
```

```python
test['Week']= pd.to_datetime(test['Week'])
(ggplot(test, aes(x="Week", y="Number", group= "Lineage")) 
+geom_line(aes(color= "Lineage"))
+scale_x_date()
+labs(title= "Weekly Infections")
+theme(figure_size=(16, 8))
)
```

```python
vaccine=pd.read_csv("vaccine_by_race_percent.csv")
```

```python
vaccine['vaccine_date']= pd.to_datetime(vaccine['vaccine_date'])
(ggplot(vaccine, aes(x= "vaccine_date", y= "% of Race/Ethnicity Vaccinated (.5+)", color= "Race/Ethnicity"))
+geom_line()
+scale_x_date()
+labs(title= "Percent Vaccinated", x= "Date", y= "Percent Vaccinated")
+theme(figure_size=(16, 8))
)
```

```python
cases_per= pd.read_csv("ethnicity.csv")
cases_per['Report Date']= pd.to_datetime(cases_per['Report Date'])
(ggplot(cases_per, aes(x= "Report Date", y= "Rate per 100k", color= "race_ethnicity"))
+geom_line()
+scale_x_datetime()
+labs(title= "Rates per 100k by Report Date", x= "Date", y= "Case Rate per 100k")
+theme(figure_size=(16, 8))
)
```

```python
death_percent= pd.read_excel("death_rate.xlsx")
death_percent['Date']= pd.to_datetime(death_percent['Date'])
(ggplot(death_percent, aes(x= "Date", y= "Death.percent"))
+geom_line(color= 'red')
+scale_x_datetime()
+labs(title= "Death Rate of Hospitalized Patients", x= "Date", y= "Percentage of Deaths in Hospitalized")
+theme(figure_size=(16, 8))
)
```

```python
(ggplot(death_percent, aes(x= "Date", y= "Cumulative.Deaths"))
+geom_line(color= 'red')
+scale_x_datetime()
+labs(title= "Cumulative Deaths Over Time", x= "Date", y= "Cumulative Deaths")
+theme(figure_size=(16, 8))
)
```
